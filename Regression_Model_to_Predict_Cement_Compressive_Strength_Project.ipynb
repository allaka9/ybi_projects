```python
# import library
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import data
cement = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Concrete%20Compressive%20Strength.csv')
```

```python
# view data
cement.head()
```

```python
# info of data
cement.info('info')
```

```python
# summary statistics
cement.describe()
```

```python
# check for missing value
cement.isna().sum()
```

```python
# check for categories
cement.nunique()
```

```python
# visualize pairplot
import seaborn
seaborn.pairplot(cement)
```

```python
# correlation
cement.corr()
```

```python
# columns name
cement.columns
```

```python
# define y
y = cement['Cement (kg in a m^3 mixture)']
```

```python
# define X
x = cement.drop('Cement (kg in a m^3 mixture)', axis=1)
```

```python
# split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.7, random_state=2529)
```

```python
# verify shape
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# train model
model.fit(x_train,y_train)
```

```python
# predict with model
y_pred=model.predict(x_test)

"""## model evaluation"""
```

```python
# model accuracy
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
# model MAE
mean_absolute_error(y_test,y_pred)
```

```python
# model MAPE
mean_absolute_percentage_error(y_test,y_pred)
```

```python
# model MSE
mean_squared_error(y_test,y_pred)
```
